```python
dframe = pd.DataFrame(car_details)
with open("data.csv","w") as f:
  dframe.to_csv(path_or_buf=f)
```

```python
# dir(dframe)
def get_per_km_cost(row):
  if row['fuel_type'] == 'petrol':
    return row["petrol_cost"] / row["fuel_mileage"]
  elif row['fuel_type'] == 'diesel':
    return row["diesel_cost"] / row["fuel_mileage"]
  else:
    return ( row["unit_cost"] * row["ev_capacity"] ) / row["ev_range"]



# total running in Month
dframe["total_running_in_Month"]  = (dframe["daly_running"] * 30) + dframe['optional_running']

# per km cost 
dframe["per_km_cost"] = dframe.apply(get_per_km_cost, axis = 1)

# yearly cost (8 to 10 years)
for each in range(1, how_many_year_of_calculation + 1):

  def get_yearly_cost(row):
    if row['fuel_type'] == 'petrol':
      inflation_rate_of_fuels_of_that_year = row['inflation_rate_of_fuels'] * each
      petrol_cost_of_that_year_per =  100 + inflation_rate_of_fuels_of_that_year
      petrol_cost_of_that_year = ( row["petrol_cost"] / 100 ) * petrol_cost_of_that_year_per

      per_km_cost = petrol_cost_of_that_year / row["fuel_mileage"]
    
    elif row['fuel_type'] == 'diesel':
      inflation_rate_of_fuels_of_that_year = row['inflation_rate_of_fuels'] * each
      diesel_cost_of_that_year_per =  100 + inflation_rate_of_fuels_of_that_year
      diesel_cost_of_that_year = ( row["diesel_cost"] / 100 ) * diesel_cost_of_that_year_per

      per_km_cost =  diesel_cost_of_that_year / row["fuel_mileage"]

    else:
      inflation_rate_of_fuels_of_that_year = row['inflation_rate_of_fuels'] * each
      unit_cost_of_that_year_per =  100 + inflation_rate_of_fuels_of_that_year
      unit_cost_of_that_year = ( row["unit_cost"] / 100 ) * unit_cost_of_that_year_per

      ev_capacity_reduction_rate_of_that_year = ev_capacity_reduction_rate * (each - 1)
      ev_capacity_reduction_rate_of_that_year_value = ( row["ev_range"] / 100) * ev_capacity_reduction_rate_of_that_year

      per_km_cost = ( unit_cost_of_that_year * row["ev_capacity"] ) / (row["ev_range"] - ev_capacity_reduction_rate_of_that_year_value)

    result = round( row["total_running_in_Month"] * 12 * per_km_cost,2 ) 
    return result


  dframe[str(each) + " year"] = dframe.apply(get_yearly_cost,axis=1)

dframe["deprication_cost_8_year_after"] = (dframe['showroom_price'] /100) * dframe['car_depreciation'] * 8

dframe["resale_value"] = dframe['on_road_price'] - dframe["deprication_cost_8_year_after"]

dframe["runining_after_that_mach_years"] = (dframe['total_running_in_Month'] * 12) * how_many_year_of_calculation

dframe["total_cost_of_car_after_difined_years"] =(dframe['on_road_price'] + dframe['1 year'] + dframe['2 year'] + dframe['3 year'] + dframe['4 year'] + dframe['5 year'] + dframe['6 year'] + dframe['7 year'] + dframe['8 year'] + dframe['9 year'] + dframe['10 year'] ) - dframe["resale_value"]
 
# + dframe['7 year'] + dframe['8 year']
# resale value after (8 to 10 years)

dframe["monthly_cost"] = dframe["total_running_in_Month"] * dframe['per_km_cost']
```

```python
def add_on_road_price(col_name,index):
  years = iter(range(1,len(df)))
  df.loc[3, col_name] = df.loc[3, col_name] + dframe['on_road_price'][index] 
  for i in range(4,len(df)+3):
    df.loc[i, col_name] = df.loc[i, col_name] + df.loc[i-1, col_name] 

pd.options.mode.chained_assignment = None
for index,each in enumerate(df.columns):
  add_on_road_price(each,index)
print(dframe['on_road_price'].to_list())
df
```

```python
import math

min_car =  dframe[dframe['on_road_price'] == min(dframe["on_road_price"])]

min_car_on_road_price =  min_car['on_road_price']

def simple_interest(p,n,i):
  # return p * (math.pow((1 + i),n) - 1)
  return (p * i * n)/100

# = [P (1 + i)n]  P
# = P [(1 + i)n  1]
# Where:

# P = principal
# i = nominal annual interest rate in percentage terms
# n = number of compounding periods

def add_intrest_on_extra_cost(col_name,fuel_type):
  on_road_price = float(dframe.iloc[fuel_type]['on_road_price'])
  for i in range(0,len(df)):
    actual_time = i
    # actual_principal = float(on_road_price - min_car_on_road_price)
    actual_rate = float(intrest_on_savings)
    df.loc[i, col_name] = df.loc[i, col_name] + simple_interest(on_road_price,actual_time,actual_rate)

fuel_types = iter(dframe.index)
for each in df.columns:
  add_intrest_on_extra_cost(each,next(fuel_types))

df
```
